```python
import nba_py
import datetime
import time
import nba_py.game
import nba_py.player
import nba_py.team
import pandas as pd
import numpy as np
import functools
import time
import pymysql
from sqlalchemy import create_engine
%matplotlib inline
```

```python
conn = create_engine('mysql+pymysql://root:@localhost:3306/nba_stats')

try:
    game_header = pd.read_sql_table('game_header', conn)
    print(len(game_header) + ' games loaded.')
except ValueError:
    print('no table yet!')
    begin = datetime.date(2012, 10, 30)
    game_header_new = nba_py.Scoreboard(month = begin.month, 
                                        day = begin.day, 
                                        year = begin.year, league_id = '00', offset = 0).game_header()

end = datetime.date.today()

for i in range((end - begin).days + 1):
    day = begin + datetime.timedelta(days = i)
    game_header_new = game_header_new.append(nba_py.Scoreboard(month = day.month, 
                                                               day = day.day, 
                                                               year = day.year, 
                                                               league_id = '00', 
                                                               offset = 0).game_header())
    print(str(day) + ' finished!')

game_header_new
```
